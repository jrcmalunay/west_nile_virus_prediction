```python
station_1_weather = weather[weather['Station']==1]
```

```python
merged_s1_test = pd.merge(station_1_weather,test,how='right', on='Date')
```

```python
merged_s1_test.WetBulb = merged_s1_test.WetBulb.map(lambda x: np.NaN if x=='M' else x)
merged_s1_test.dropna(inplace=True)
merged_s1_test.WetBulb.value_counts()
merged_s1_test.WetBulb = merged_s1_test.WetBulb.map(lambda x: int(x))
merged_s1_test.Heat = merged_s1_test.Heat.map(lambda x: int(x))
merged_s1_test['Tavg'] = merged_s1_test['Tavg'].map(lambda x: float(x))
merged_s1_test['Depart'] = merged_s1_test['Depart'].map(lambda x: float(x))
merged_s1_test['AvgSpeed'] = merged_s1_test['AvgSpeed'].map(lambda x: float(x))
merged_s1_test['Heat'] = merged_s1_test['Heat'].map(lambda x: float(x))
merged_s1_test['Cool'] = merged_s1_test['Cool'].map(lambda x: float(x))
merged_s1_test['Depth'] = merged_s1_test['Depth'].map(lambda x: float(x))
merged_s1_test['ResultDir'] = merged_s1_test['ResultDir'].map(lambda x: float(x))
merged_s1_test['StnPressure'] = merged_s1_test['StnPressure'].map(lambda x: float(x))
merged_s1_test['SeaLevel'] = merged_s1_test['SeaLevel'].map(lambda x: float(x))
```

```python
merged_s1_test.drop('Water1', axis=1, inplace=True)
merged_s1_test.drop('SnowFall', axis=1, inplace=True)
merged_s1_test.drop('Depth',axis=1, inplace=True)
```

```python
#Convert Date time
merged_s1_test.Date = pd.to_datetime(merged_s1_test.Date)
```

```python
#PrecipTotal, '  T' "Trace Amounts" = 0.0
merged_s1_test.PrecipTotal = merged_s1_test.PrecipTotal.map(lambda x: 0.00 if x=='  T' else x)
merged_s1_test.PrecipTotal = merged_s1_test.PrecipTotal.map(lambda x: float(x))
```

```python
import patsy
from patsy import dmatrices

formula = 'WnvPresent ~ Tmax + Tmin + Tavg + Depart + DewPoint + WetBulb + Heat + Cool + C(Sunrise) + C(Sunset) + C(CodeSum)\
           + PrecipTotal + StnPressure + SeaLevel + ResultSpeed + ResultDir + AvgSpeed + C(Address) + C(Species) + Block\
           + C(Street) + C(Trap) + C(AddressNumberAndStreet) + Latitude + Longitude + AddressAccuracy + NumMosquitos -1'

formula = str(formula)

Yb, Xb = dmatrices(formula, data = merged_s1_train, return_type = 'dataframe')
```

```python
Xb_mat = (Xb - Xb.mean())/Xb.std()
```

```python
from sklearn.metrics import accuracy_score


logreg = LogisticRegression()

model = logreg.fit(Xb_mat, Yb)

predictions = logreg.predict(Xb_mat)

print accuracy_score(Yb, predictions)
```
